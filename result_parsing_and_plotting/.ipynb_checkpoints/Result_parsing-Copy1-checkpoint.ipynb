### Double Typing
Merge rows, where a person has been typed twice.

```python
print(len(set(list(MG_exome_df['id']))))
print(len(list(MG_exome_df['id'])))
```

```python
non_unique = list({x for x in list(MG_exome_df['id']) if list(MG_exome_df['id']).count(x) > 1})
non_unique_df = MG_exome_df[MG_exome_df['id'].isin(non_unique)]
non_unique_df
```

```python
clean_duplicates_df = pd.DataFrame()
for column in ['A','A.1','B','B.1','C','C.1','DRB1','DRB1.1','DQB1','DQB1.1']:
    clean_duplicates_df[column] = non_unique_df.groupby(['id','sbgroup'])[column].apply('/'.join)

clean_duplicates_df.reset_index(inplace=True)
```

```python
#Remove the duplicate columns from the full dataframe"
MG_exome_df = MG_exome_df[~MG_exome_df['id'].isin(non_unique)]

#Add back the clean duplicate rows:
MG_exome_df.append(clean_duplicates_df, sort=False)
```

```python
print(len(set(list(MG_exome_df['id']))))
print(len(list(MG_exome_df['id'])))
```

```python
MG_exome_df.reset_index(inplace=True)
del MG_exome_df['index']
```

### Ambiguity

```python
#Find out, how many have ambiguous typing at 4-field resolution. 
#This is however not reason for discarding ambiguous data, as ambigouity only is noted at places where
#exome 2 and 3 (HLA class I) and exome 2 (HLA class II) are identical between the alleles noted

trash = 0
treasure = 0

treasure_list = list()

#Edit table to remove all entries, where ambiguity exists in the two field resolution:
for row in mille_gs_df.iterrows():
    hit = False
    for entry in list(row[1]):
        len(set([i[0:5] for i in entry.split('/')]))
        
        if len(set([i[0:5] for i in entry.split('/')])) != 1:
            hit = True
            trash += 1
            #print(row[1][0])

    if hit == False:
        treasure += 1
        #print(row[1][0])
        treasure_list.append(row[1][0])

print("treasure:", treasure)
print("trash:", trash)
```

# Result checking ATHLATES dataset

FIND BETTER WAY OF CHECKING RESULTS
Make sure gold standard = 0000 doesn't result in an error.

```python
import re

n_tot = 0
kourami_score = 0
hla_la_score = 0
optitype_score = 0
hisatgenotype_score = 0

#Loop over all alleles:
for i in gs_df[gs_df['Subject'] == 'HG01756'].iloc[0][3:]:
    facit = re.match(r'^\w+\*\d{2}:\d{2}', i).group(0)
    n_tot += 1
    print(facit)
    
    #Need code to check for duplicates and for missing input e.g. one allele missing from gold standard 
    
    #Check if kourami predicted this allele:
    for prediction_kourami in list(kourami_results):
        if facit in prediction_kourami:
            print("Kourami", prediction_kourami)
            kourami_score += 1
            break
    
    #Check if HLA_LA predicted this allele:
    for prediction_hlala in list(hla_la_results):
        if facit in prediction_hlala:
            print("HLA-LA", prediction_hlala)
            hla_la_score += 1
            break
            
    #Check if optitype predicted this allele:
    for pred_optitype in list(optitype_results):
        if facit in pred_optitype:
            print("Optitype:", pred_optitype)
            optitype_score += 1
            break
    
    #Check if hisatgenotype predicted this allele:
    for pred_hisatgenotype in list(hisatgenotype_resultlist):
        if facit in pred_hisatgenotype:
            print("hisatgenotype:", pred_hisatgenotype)
            hisatgenotype_score += 1
            break
```

```python
print("Kourami prediction accuracy:", kourami_score/n_tot*100, "%")
print("HLA-LA prediction accuracy:", hla_la_score/n_tot*100, "%")
print("Optitype prediction accuracy:", optitype_score/n_tot*100, "%")
print("Hisatgenotype prediction accuracy:", hisatgenotype_score/n_tot*100, "%")
```
